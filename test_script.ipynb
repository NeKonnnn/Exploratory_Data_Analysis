```python
from eda_script import EDAProcessor
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_1.csv', delimiter=';')

# Преобразование колонки Dt_Customer в формат datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d.%m.%Y')

df.head(1)
```

```python
eda = EDAProcessor(df)

# 1. Сводная таблица пропусков
eda_summary = eda.generate_eda_summary()
# Вывод таблицы
display(eda_summary['summary'])

# Вывод количества дубликатов строк
print(f"Всего дубликатов строк: {eda_summary['duplicate_count']}")

# Вывод самих дубликатов строк
if not eda_summary['duplicates'].empty:
    print("Дублированные строки:")
    display(eda_summary['duplicates'])
else:
    print("Дубликатов строк не найдено.")

# Вывод дублированных столбцов
if eda_summary['duplicate_columns']:
    print("Дублированные столбцы:")
    for col1, col2 in eda_summary['duplicate_columns']:
        print(f"- {col1} и {col2}")
else:
    print("Дублированных столбцов не найдено.")

# 2. Распределение целевой переменной
target_summary = eda.plot_target_distribution('Response') # Поменять название таргета
display(target_summary)

# 3. Определяем категориальные и числовые столбцы автоматически
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
numeric_columns = df.select_dtypes(include=['number']).columns.tolist()

# 4. Распределение категориальных переменных
categorical_summary = eda.plot_categorical_distributions(categorical_columns)
display(categorical_summary)

# 5. Сравнение категориальных переменных с целевой переменной
eda.plot_categorical_vs_target('Response', categorical_columns) # Поменять название таргета

# 6. Построение гистограмм и KDE
kde_summary = eda.plot_kde_distributions(numeric_columns)
display(kde_summary)

# 7. Анализ выбросов на основе IQR
if not numeric_columns:
    print("Нет числовых столбцов для анализа выбросов (IQR).")
    outlier_summary = pd.DataFrame()
else:
    outlier_summary = eda.detect_outliers_iqr(numeric_columns)
    display(outlier_summary)

# 8. Анализ выбросов на основе Z-score
if not numeric_columns:
    print("Нет числовых столбцов для анализа выбросов (Z-score).")
    zscore_outliers = pd.DataFrame()
else:
    zscore_outliers = eda.detect_outliers_zscore(numeric_columns, threshold=3)
    display(zscore_outliers)

# 9. Анализ таблиц сопряженности
cross_tabulations, cross_tab_conclusions = eda.analyze_categorical_cross_tabulations(categorical_columns)

# Отображение объединённой таблицы сопряженности
display(cross_tabulations)

# Печать выводов
for pair, conclusion in cross_tab_conclusions.items():
    print(f"{pair}: {conclusion}")
    
# 10. Выявление редких категорий
rare_categories = eda.find_rare_categories(categorical_columns, threshold=0.05)
rare_categories_df = pd.DataFrame(
    [{"Признак": col, "Редкие категории": ", ".join(vals)} for col, vals in rare_categories.items()]
)
display(rare_categories_df)

# # 11. Анализ линейной зависимости между числовыми признаками
# eda.plot_numeric_pairplot()

# 12. Анализ корреляций

# Фильтруем только числовые столбцы, включая таргет, если он числовой
numeric_df = df.select_dtypes(include=['number'])

# Инициализируем EDAProcessor только для числовых данных
eda_numeric = EDAProcessor(numeric_df)

# Выполняем анализ корреляций с таргетом, если он числовой
correlation_result = eda_numeric.analyze_correlations(target_column='Response', threshold=0.5) # Поменять название таргета

# Вывод корреляций числовых признаков с таргетом
if 'correlations' in correlation_result:
    print("Корреляции числовых признаков с таргетом:")
    display(correlation_result['correlations'])

# С категориальным таргетом =
# results = eda.analyze_correlations(target_column='Category1', threshold=0.5)

# Cramer's V для категориального таргета 
# if 'cramers_v' in results:
#     print("Критерий Крамера V:")
#     display(results['cramers_v'])

# Тест ANOVA для категориального таргета 
# if 'anova' in results:
#     print("Тест ANOVA для категориального таргета:")
#     display(results['anova'])

# 13. Анализ временных атрибутов
time_intervals_summary = eda.analyze_datetime_attributes('Dt_Customer')  # Укажите столбец с датами

# Отображение результата
for attr, table in time_intervals_summary.items():
    print(f"Сводная таблица для '{attr}':")
    display(table)

# 14. Декомпозиция временного ряда
decomposition_stats = eda.decompose_time_series('Dt_Customer', 'MntWines')  # Укажите столбец с датами и значениями
print("Статистики декомпозиции:")
display(decomposition_stats)

# 15. Построение автокорреляции и частичной автокорреляции
acf_pacf_table = eda.plot_autocorrelations('Dt_Customer', 'MntWines', lags=50)
display(acf_pacf_table)

# 16. Проверка стационарности временного ряда
stationarity_result = eda.check_stationarity('Dt_Customer', 'MntWines')  # Укажите столбец с датами и значениями
print("Результаты теста Дики-Фуллера:")
display(stationarity_result)

# 17. Построение тепловой карты сезонности
seasonality_pivot = eda.plot_seasonality_heatmap('Dt_Customer', 'MntWines', freq='month')  # Укажите столбец с датами и значениями
print("Используемая таблица для тепловой карты:")
display(seasonality_pivot)

# 18. Построение временного ряда с таблицей
time_series_table = eda.plot_time_series_with_table('Dt_Customer', 'MntWines')
display(time_series_table)
```

```python
# Словарь с таблицами
summaries = {
    "Сводная таблица": eda_summary['summary'],
    "Распределение целевой переменной": target_summary,
    "Категориальные переменные": categorical_summary,
    "Выбросы IQR": outlier_summary,
    "Выбросы Z-score": zscore_outliers,
    "Редкие категории": rare_categories_df,
    "KDE Сводная таблица": kde_summary,
    "Декомпозиция временного ряда": decomposition_stats,
    "ACF_PACF": acf_pacf_table,
    "Стационарность временного ряда": stationarity_result,
    "Сезонность (Таблица)": seasonality_pivot,
    "Таблица временных рядов": time_series_table,
}

# Добавляем сводные таблицы временных интервалов
for unit, stats_df in time_intervals_summary.items():
    summaries[f"Интервалы ({unit})"] = stats_df

# Автоматически добавляем найденные дубли
for key, MntWines in {
    "Дублированные строки": eda_summary['duplicates'],
    "Дублированные столбцы": pd.DataFrame(eda_summary['duplicate_columns'], columns=["Столбец 1", "Столбец 2"])
}.items():
    if not MntWines.empty:  # Проверяем, что DataFrame не пустой
        summaries[key] = MntWines

# Если есть корреляции, добавляем их в словарь
if 'correlations' in correlation_result:
    summaries["Корреляции числовых признаков"] = correlation_result['correlations']

# Сохранение в Excel
eda.save_all_summaries_to_excel(summaries)
```
